# MoMA Explore

This is a collection of code for my exploration of data provided by MoMA.
This data is associated with a Kaggle competition which can be found [here](https://www.kaggle.com/momanyc/museum-collection/data).

```python
from bokeh.plotting import figure, output_notebook, show
import pandas as pd
import numpy as np
output_notebook()
```

```python
# Reads in the data and fixes the index for each frame
artworks =  pd.read_csv("../../data_production/museum-collection/artworks_sample.csv", encoding="utf-8")
artists = pd.read_csv("../../data_production/museum-collection/artists_sample.csv", encoding="utf-8")
artists = artists.reset_index(drop=True)
artworks = artworks.reset_index(drop=True)
```

```python
# Create extra columns
artworks.loc[:, "Area"] = None # Creates Column of Area
artists.loc[:,"Average"] = None # Creates Column of Average
```

```python
x = artists["Artist ID"]
y = artists["Birth Year"]
p = figure(title="Artist ID by Birth", x_axis_label='ID', y_axis_label='Birth Year')

p.line(x, y, legend="Temp.", line_width=2)


show(p, notebook_handle=True)
```

In the next group of code we will calculate the average work size for each artist, if the required data is provided.

```python
# print(artists["Artist ID"])
for index, row in artists.iterrows():
    temp = artworks.loc[str(row["Artist ID"]) == artworks["Artist ID"]]
    if len(temp) > 0:
        temp_total = 0
        for i, piece in temp.iterrows():
            area = piece["Height (cm)"] * piece["Width (cm)"]
#             print(piece["Height (cm)"])
            piece["Area"] = area
            artworks.loc[piece["Artwork ID"] == artworks["Artwork ID"], ["Area"]] = area
            temp_total = temp_total + area
        avg = temp_total / len(temp)
#         print("AVG Area: ", avg)
        artists.loc[index, "Average"] = avg


artworks.to_csv("output_artworks.csv")
artists.to_csv("output_artists.csv")
```
